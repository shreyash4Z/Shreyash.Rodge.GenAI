Q:5 Generate a model for Covid 19 with symptoms of parameters like fever, cold, shivering,
weight loss, generate 100 model data with random values for each parameter and order by
parameter lowest to highest in display based on the input parameter.

```
import numpy as np
import pandas as pd
```

```
parameters = ['fever', 'cold', 'shivering', 'weight_loss']
num_samples = 100
```

```
data = np.random.randint(0, 11, size=(num_samples, len(parameters)))
df = pd.DataFrame(data, columns=parameters)
```

```
def sort_by_parameter(df, parameter):
    if parameter in parameters:
        return df.sort_values(by=parameter)
    else:
        return f"Invalid parameter. Choose from {parameters}"
```

```
pd.set_option('display.max_rows', None)
parameter_to_sort = input(f"Sort by one of {parameters}: ")
sorted_df = sort_by_parameter(df, parameter_to_sort)
print(sorted_df)
```

Q:3 Generate a model for an Insurance company to hold information on the insurer's vehicle,
and create a chart of monthly, yearly, and qtrly premiums based on no. of years of insurance
where in each year, the value of the vehicle depreciates by 7%.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
initial_value = float(input("Enter the initial value of the vehicle: "))
num_years = int(input("Enter the number of years of insurance: "))
premium_rate = float(input("Enter the premium rate as a percentage (e.g., 5 for 5%): ")) / 100
```

```
def calculate_depreciated_value(initial_value, year, depreciation_rate=0.07):
    return initial_value * (1 - depreciation_rate) ** year
```

```
years = np.arange(1, num_years + 1)
depreciated_values = [calculate_depreciated_value(initial_value, year) for year in years]
yearly_premiums = [value * premium_rate for value in depreciated_values]
quarterly_premiums = [premium / 4 for premium in yearly_premiums]
monthly_premiums = [premium / 12 for premium in yearly_premiums]
```

```
df = pd.DataFrame({
    'Year': years,
    'Depreciated Value': depreciated_values,
    'Yearly Premium': yearly_premiums,
    'Quarterly Premium': quarterly_premiums,
    'Monthly Premium': monthly_premiums
})
print(df)
```

```
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Yearly Premium'], label='Yearly Premium', marker='o')
plt.plot(df['Year'], df['Quarterly Premium'], label='Quarterly Premium', marker='s')
plt.plot(df['Year'], df['Monthly Premium'], label='Monthly Premium', marker='^')

plt.title('Premiums Over Time Based on Depreciation')
plt.xlabel('Year')
plt.ylabel('Premium Amount')
plt.legend()
plt.grid(True)
plt.show()
```
